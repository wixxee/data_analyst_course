## Описание проекта

### Задача
Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных.<br>
Распространённая проблема фитнес-клубов и других сервисов — отток клиентов. Как понять, что клиент больше не с вами? Можно записать в отток тех, кто попросил закрыть договор или удалил аккаунт. Однако клиенты не всегда уходят демонстративно: чаще перестают пользоваться сервисом тихо.<br>
Индикаторы оттока зависят от специфики отрасли. Когда пользователь редко, но стабильно закупается в интернет-магазине — не похоже, что он «отвалился». А вот если две недели не заходит на канал с ежедневно обновляемым контентом, дела плохи: подписчик заскучал и, кажется, оставил вас.<br>
Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал. Конечно, не исключено, что он уехал на Бали и по приезде обязательно продолжит ходить на фитнес. Однако чаще бывает наоборот. Если клиент начал новую жизнь с понедельника, немного походил в спортзал, а потом пропал — скорее всего, он не вернётся.<br>
Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество клиентских анкет. Ваша задача — провести анализ и подготовить план действий по удержанию клиентов.<br>

А именно:<br>
научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;<br>
сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;<br>
проанализировать основные признаки, наиболее сильно влияющие на отток;<br>
сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:<br>
1) выделить целевые группы клиентов;<br>
2) предложить меры по снижению оттока;<br>
3) определить другие особенности взаимодействия с клиентами.<br>

### Описание данных
**Данные клиента за предыдущий до проверки факта оттока месяц:**
- `gender` — пол;
- `Near_Location` — проживание или работа в районе, где находится фитнес-центр;
- `Partner` — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);
- `Promo_friends` — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента);
- `Phone` — наличие контактного телефона;
- `Age` — возраст;
- `Lifetime` — время с момента первого обращения в фитнес-центр (в месяцах).

**Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:**
- `Contract_period` — длительность текущего действующего абонемента (месяц, 6 месяцев, год);
- `Month_to_end_contract` — срок до окончания текущего действующего абонемента (в месяцах);
- `Group_visits` — факт посещения групповых занятий;
- `Avg_class_frequency_total` — средняя частота посещений в неделю за все время с начала действия абонемента;
- `Avg_class_frequency_current_month` — средняя частота посещений в неделю за предыдущий месяц;
- `Avg_additional_charges_total` — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.

**`Churn` — факт оттока в текущем месяце.**

## Построение модели оттока клиентов

```python
#выделим признаки и разделим данные для построения и проверки моделей
y = df['churn']
x = df.drop('churn', axis = 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=0)
```

```python
# обучим модель и проверим метрики
models = [LogisticRegression(random_state=0), RandomForestClassifier(random_state=0)]
for i in models:
    model = i
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(
        'accuracy:{:.2f} precision:{:.2f} recall:{:.2f} '.format(
            accuracy_score(y_test, y_pred),
            precision_score(y_test, y_pred),
            recall_score(y_test, y_pred)
        )
    )
```

- модель логической регрессии показывает лучший результат.

## Кластеризация клиентов

```python
#стандартизируем данные
sc = StandardScaler()
x_sc = sc.fit_transform(x)
```

```python
#постройте матрицу расстояний функцией linked
linked = linkage(x_sc, method = 'ward')
```

```python
#визуализируем дендрограмму
plt.figure(figsize=(15, 10))
dendrogram(linked, orientation='top')
plt.title('Hierarchial clustering')
plt.show()
```

- оптимальное число кластеров - 5;

```python
#выполним кластеризацию методом K-Means
km = KMeans(n_clusters=5, random_state=0)
labels = km.fit_predict(x_sc)
df['cluster'] = labels

print(df.groupby('cluster').mean())
```

- чем больше контрактный период, человек потратил на занятия, больше групповых тренировок, тем выше шанс того, что клиент останется;
- 1 кластер наиболее надежный - чаще всего это сотрудник компании партнера, чуть старше 29, в близкой локации, договором длиннее полугода и в среднем около 2 посещений, в половине случаев посещяет групповые занятия, траты в клубе наибольшие;
- 0 кластер наименее надежен - младше 29, срок договора около 3 месяцев, далеко от клуба, и только в половине случаев сотрудник компании партнера, среднее количество посещений 1.8, групповые занятия выбираются в четверти случаях, траты в клубе наименьшие;
- 2 кластер имеет схожие показатели оттока, но что интересно - партнер клуба уже в четверти случаях, при близком расположении, так же не посещаются занятия в группе;
- 3 и 4 кластер имеют общие черты, однако различие в партнертсве клуба, в 3 кластере посетитель не партнер клуба, но отток ниже, похоже дело в параметре групповых занятий, которые всегда посещяются данным кластером.

- групповые занятия тоже хорошо удерживают клиентов - это, кажется, логичным, групповые занятия с одной стороны, дают больше социального клея, который удерживает клиента, с другой дают большое разнообразие в тренировках;
- так же, потраченные деньги, тоже коррелируют с удержанием, поэтому данное направление тоже необходимо усиливать;
- безусловно, необходимо не забывать об условных группах рисков клиентов - более молодых и тех, у кого уже заканчивается абонимент;
- так что стоит попытаться увеличить срок абонемента, например с помощью различных промоакций, что в краткосрочной перспективе, конечно, немного повысит косты, но в долгосрочной - удержит клиента и позволит продолжить зарабатывать на различный услугах.
