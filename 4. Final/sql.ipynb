## Выполнение заданий

### Посчитайте, сколько книг вышло после 1 января 2000 года;

```python
query = ''' SELECT COUNT(book_id) as numbers_of_books
            FROM books 
            WHERE publication_date > '2000-01-01'
        '''
```

```python
pd.io.sql.read_sql(query, con = engine)
```

После 01.01.2020 вышло 819 книг.

### Для каждой книги посчитайте количество обзоров и среднюю оценку;

```python
query = ''' SELECT b.book_id,
                   MAX(b.title),
                   AVG(r.rating) as avg_rating,
                   COUNT(rev.review_id) as number_of_review                   
            FROM books as b
            LEFT JOIN ratings as r on b.book_id=r.book_id
            LEFT JOIN reviews as rev ON r.book_id=rev.book_id
            GROUP BY b.book_id
            ORDER BY number_of_review DESC
        '''
```

```python
pd.io.sql.read_sql(query, con = engine).head(10)
```

### Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры;

```python
query = ''' SELECT p.publisher,
                   COUNT(b.book_id) as number_of_books                   
            FROM books as b
            JOIN publishers as p on b.publisher_id=b.publisher_id
            WHERE b.num_pages > 50
            GROUP BY p.publisher
            ORDER BY COUNT(b.book_id) DESC
            LIMIT 1
        '''
```

```python
pd.io.sql.read_sql(query, con = engine).head()
```

### Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками;

```python
query = ''' WITH
            r AS (SELECT book_id,
                         COUNT(rating_id) as rating_count,
                         AVG(rating) as avg_rating
                  FROM ratings
                  GROUP BY book_id)
            SELECT a.author,
                   AVG(r.avg_rating)                 
            FROM r
            JOIN books as b on r.book_id=b.book_id
            JOIN authors as a on a.author_id=b.author_id 
            WHERE r.rating_count>50
            GROUP BY a.author
            ORDER BY AVG(r.avg_rating) DESC
            LIMIT 1
        '''
```

```python
pd.io.sql.read_sql(query, con = engine).head()
```

### Посчитайте среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
query = ''' WITH
            rat AS (SELECT username,
                           COUNT(rating_id) as rating_count
                  FROM ratings
                  GROUP BY username),
            rev AS (SELECT username,
                           COUNT(review_id) as rating_count
                  FROM reviews
                  GROUP BY username)
            SELECT AVG(rev.rating_count) AS avg_reviews               
            FROM rat
            JOIN rev on rat.username=rev.username
            WHERE rat.rating_count>50
        '''
```

```python
pd.io.sql.read_sql(query, con = engine).head()
```
